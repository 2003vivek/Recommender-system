```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_500_credits.csv')
```

```python
movie=movies.merge(credits,on=credits.title)
movie=movie.drop(['title_y','title_x'],axis=1)
movie.head(1)
```

```python
movies=movie[['key_0','genres','id','keywords','overview','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
movies=movies.dropna()
movies.isnull().sum()
```

```python
movies.duplicated().any()
```

```python
import ast
# ast.literal_eval(node_or_string=movies['genres'])
def convert(obj):
    li=[]
    for i in ast.literal_eval(obj):
        li.append(i['name'])
    return li
```

```python
movies['genres']=movies['genres'].apply(convert)
movies
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies
```

```python
def convert3 (obj):
    li=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            li.append(i['name'])
            counter+=1
    return li
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies
```

```python
def get_director(obj):
    li=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            li.append(i['name'])
            break
    return li
            
```

```python
movies['crew']=movies['crew'].apply(get_director)
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
movies
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
movies.head(1)
```

```python
df=movies[['key_0','id','tags']]
df
```

```python
df['name']=df['key_0']
df
```

```python
new_df=df[['name','id','tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
new_df['name'].unique()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    li=[]
    for i in text.split():
        li.append(ps.stem(i))
    return " ".join(li)
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer(max_features=5000,stop_words='english')
vector=v.fit_transform(new_df['tags']).toarray()
v.get_feature_names()
```

```python
new_df
```

```python
v.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cs=cosine_similarity(vector) # 1 means movies are similar and o means dissimilar
```

```python
cs[2] # HERE THIS IS THE DISTANCE OF A PARTICULAR MOVIE FROM ALL THE MOVIES
```

```python
cs.shape
```

```python
new_df.loc[1216]['name']
```

```python
def recommend(movie):
    movie_index=new_df[new_df['name']==movie].index[0]
    movie_distance=cs[movie_index]
    sortlist=sorted(enumerate(cs[movie_index]),reverse=True,key=lambda x: x[1])[1:6]
    
    for i in sortlist:
        recommendations=new_df.loc[i[0]]['name']
        print(recommendations)
        
```

```python
recommend('Batman')
```

```python
recommend('John Carter')
```
